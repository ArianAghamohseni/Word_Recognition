```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def word_drift(k1, beta, frequency, alpha, word_prob):
    return (k1 + beta * frequency)/(1 + np.exp(-alpha * (word_prob - 0.5)))

def nonword_drift(k2, alpha, nonword_prob):
    return k2/(1 + np.exp(-alpha * (nonword_prob - 0.5)))
```

### Sigmoid Function behavior

```python
plt.figure(figsize = (15, 5), dpi=100)

x = np.linspace(0, 5, 10)
y1 = word_drift(1, 0, 0, 1, x)
y2 = word_drift(2, 0, 0, 1, x)
y3 = word_drift(0.5, 0, 0, 1, x)
plt.plot(x, y1, color='tab:blue', label="K=1")
plt.plot(x, y2, color='tab:orange', label="K=2")
plt.plot(x, y3, color='tab:red', label="K=0.5")

# Add features to our figure
plt.legend()
plt.grid(True, linestyle =':')

plt.title('Sigmoid function with different Asymptote')
plt.xlabel('X')
plt.ylabel('Y')
plt.savefig("Sigmoid.pdf")
```

```python
prob = np.linspace(0, 1, 10)

w_param_set_1 = [1, 0.5, 0.6]
w_param_set_2 = [1, 0.8, 0.6]
w_param_set_3 = [2, 0.5, 0.6]
w_param_set_4 = [1, 0.5, 0.3]

nw_param_set_1 = [1, 0.5]
nw_param_set_2 = [1, 0.8]
nw_param_set_3 = [2, 0.5]
```

### Sigmoid Function with constant frequency

```python
fig, axs = plt.subplots(1, 3,
                        figsize = (20, 8), dpi=100,
                        sharex=True)


word_drift_set_LF_1 = word_drift(w_param_set_1[0], w_param_set_1[1],
                              2, w_param_set_1[2], prob)
word_drift_set_HF_1 = word_drift(w_param_set_1[0], w_param_set_1[1],
                              5, w_param_set_1[2], prob)

word_drift_set_LF_2 = word_drift(w_param_set_2[0], w_param_set_2[1],
                              2, w_param_set_2[2], prob)
word_drift_set_HF_2 = word_drift(w_param_set_2[0], w_param_set_2[1],
                              5, w_param_set_2[2], prob)

word_drift_set_LF_3 = word_drift(w_param_set_3[0], w_param_set_3[1],
                              2, w_param_set_3[2], prob)
word_drift_set_HF_3 = word_drift(w_param_set_3[0], w_param_set_3[1],
                              5, w_param_set_3[2], prob)

word_drift_set_LF_4 = word_drift(w_param_set_4[0], w_param_set_4[1],
                              2, w_param_set_4[2], prob)
word_drift_set_HF_4 = word_drift(w_param_set_4[0], w_param_set_4[1],
                              5, w_param_set_4[2], prob)


non_word_drift_set_1 = nonword_drift(nw_param_set_1[0], nw_param_set_1[1],
                                     prob)
non_word_drift_set_2 = nonword_drift(nw_param_set_2[0], nw_param_set_2[1],
                                     prob)
non_word_drift_set_3 = nonword_drift(nw_param_set_3[0], nw_param_set_3[1],
                                     prob)

axs[0].plot(prob, word_drift_set_LF_1, alpha=0.6,
         label="k1=1, beta=0.5, alpha=0.6, Zipf=2")
axs[0].plot(prob, word_drift_set_LF_2, alpha=0.6,
         label="k1=1, beta=0.8, alpha=0.6, Zipf=2")
axs[0].plot(prob, word_drift_set_LF_3, alpha=0.6,
         label="k1=2, beta=0.5, alpha=0.6, Zipf=2")
axs[0].plot(prob, word_drift_set_LF_4, alpha=0.6,
         label="k1=1, beta=0.5, alpha=0.3, Zipf=2")

axs[0].legend()
axs[0].grid(True, linestyle =':')
axs[0].set_title('Drift rate value for low-frequency words')
axs[0].set_xlabel('Probability of being word')
axs[0].set_ylabel('Drift rate value')

axs[1].plot(prob, word_drift_set_HF_1, alpha=0.6,
         label="k1=1, beta=0.5, alpha=0.6, Zipf=5")
axs[1].plot(prob, word_drift_set_HF_2, alpha=0.6,
         label="k1=1, beta=0.8, alpha=0.6, Zipf =5")
axs[1].plot(prob, word_drift_set_HF_3, alpha=0.6,
         label="k1=2, beta=0.5, alpha=0.6, Zipf=5")
axs[1].plot(prob, word_drift_set_HF_4, alpha=0.6,
         label="k1=1, beta=0.5, alpha=0.3, Zipf=5")

axs[1].legend()
axs[1].grid(True, linestyle =':')
axs[1].set_title('Drift rate value for high-frequency words')
axs[1].set_xlabel('Probability of being word')
axs[1].set_ylabel('Drift rate value')

axs[2].plot(prob, non_word_drift_set_1, alpha=0.6,
         label="k2=1, beta=0.5")
axs[2].plot(prob, non_word_drift_set_2,
         label="k2=1, beta=0.8", alpha=0.6)
axs[2].plot(prob, non_word_drift_set_3,
         label="k2=2, beta=0.5", alpha=0.6)

axs[2].legend()
axs[2].grid(True, linestyle =':')
axs[2].set_title('Drift rate value for non-words')
axs[2].set_xlabel('Probability of being non-word')
axs[2].set_ylabel('Drift rate value')

plt.savefig("Mapping Functions.pdf")
```

### non-decision mapping functions

```python
plt.figure(figsize = (10, 5), dpi=100)

x = np.linspace(0, 7, 100)
y = np.exp(-x)
plt.plot(x, y, color='tab:blue')

# Add features to our figure
plt.grid(True, linestyle =':')

plt.title('Exponential function')
plt.xlabel('Frequency (Zipf: 0-7)')
plt.ylabel('Y')
plt.savefig("Exponential.pdf")
```

```python
plt.figure(figsize = (10, 5), dpi=100)

x = np.linspace(0, 7, 100)
y1 = 0.2 + 0.6 * np.exp(-x)
y2 = 0.2 + 0.3 * np.exp(-x)
y3 = 0.3 + 0.6 * np.exp(-x)
plt.plot(x, y1, color='tab:blue', label="m=0.2, gamma=0.6")
plt.plot(x, y2, color='tab:orange', label="m=0.2, gamma=0.3")
plt.plot(x, y3, color='tab:red', label="m=0.3, gamma=0.6")

# Add features to our figure
plt.legend()
plt.grid(True, linestyle =':')

plt.title('Non-deicison time mapping function')
plt.xlabel('Frequency (Zipf: 0-7)')
plt.ylabel('non-deicision time')
plt.savefig("non-decision time.pdf")
```
